## 3. Feature Engineering
### 3-1. Check Missing Values

```python
all_feature_df = train.copy().drop(columns='SalePrice').set_index('Id')
print('All columns count: ' + str(len(all_feature_df.columns)))
```

```python
all_feature_null_df = all_feature_df.isnull().sum()
all_feature_null_df = pd.DataFrame(all_feature_null_df, columns=['Missing Value Count'])
all_feature_null_df['Missing Value Ratio'] = all_feature_null_df['Missing Value Count']/train.shape[0]
missing_true_df = all_feature_null_df[all_feature_null_df['Missing Value Count']!=0]
missing_true_df.sort_values('Missing Value Count',ascending=False,inplace=True)
missing_true_df
```

```python
plt.figure(figsize=(23,12))
sns.barplot(missing_true_df.index, missing_true_df['Missing Value Ratio'])
plt.show()
print('The list of the variables which have missing values:')
print(list(missing_true_df.index))
print('\n'+'Total '+str(len(missing_true_df.index))+' variables have missing values. So let\'s fill it up.')
```

### 3-2. Missing Values Imputation

```python
modeling_data = train.copy()
modeling_data.set_index('Id',inplace=True)
```

**1. PoolQC**
- Description: Pool quality
- Processing: 'NaN'은 Pool을 갖추고 있지 않은 집이라 판단하여 'None'으로 대체

```python
modeling_data['PoolQC'] = modeling_data['PoolQC'].fillna('None')
```

**2. MiscFeature**
- Description: Miscellaneous feature not covered in other categories
- Processing: 'NaN'은 기타기능을 의미하는 이 변수에 포함되지 않는 것으로 판단하여 'None'으로 대체

```python
modeling_data['MiscFeature'] = modeling_data['MiscFeature'].fillna('None')
```

**3. Alley**
- Description: : Type of alley access
- Processing: 'NaN'은 인접한 Alley가 없는 집이라 판단하여 'None'로 대체

```python
modeling_data['Alley'] = modeling_data['Alley'].fillna('None')
```

**4. Fence**
- Description: : Fence quality
- Processing: 'NaN'은 Fence를 가지고 있지 않은 집이라 판단하여 'None'로 대체

```python
modeling_data['Fence'] = modeling_data['Fence'].fillna('None')
```

**5. FireplaceQu**
- Description: : Fireplace quality
- Processing: 'NaN'은 Fireplace를 갖고 있지 않은 집이라 판단하여 'None'로 대체

```python
modeling_data['FireplaceQu'] = modeling_data['FireplaceQu'].fillna('None')
```

**6. LotFrontage**
- Description: : Linear feet of street connected to property
- Processing: 동일한 Neighborhood 그룹의 LotFrontage에 대한 중간값으로 대체

```python
modeling_data['LotFrontage'] = \
modeling_data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

**7~11. GarageType** & **GarageYrBlt** & **GarageFinish** & **GarageQual** & **GarageCond**
- Description: : Location & Built year & Interior finish & Quality & Condition of Garage 
- Processing: 'NaN'은 Garage가 없는 집이라 판단하여 'None'로 대체

```python
modeling_data['GarageType'] = modeling_data['GarageType'].fillna('None')
modeling_data['GarageYrBlt'] = modeling_data['GarageYrBlt'].fillna('None')
modeling_data['GarageFinish'] = modeling_data['GarageFinish'].fillna('None')
modeling_data['GarageQual'] = modeling_data['GarageQual'].fillna('None')
modeling_data['GarageCond'] = modeling_data['GarageCond'].fillna('None')
```

**12~16. BsmtExposure** & **BsmtFinType1** & **BsmtFinType2** & **BsmtCond** & **BsmtQual**
- Description: : Walkout or garden level basement walls & Finished area1,2 Quality & Condition & Height of Basement
- Processing: 'NaN'은 Basement가 없는 집이라 판단하여 'None'로 대체

```python
modeling_data['BsmtExposure'] = modeling_data['BsmtExposure'].fillna('None')
modeling_data['BsmtFinType1'] = modeling_data['BsmtFinType1'].fillna('None')
modeling_data['BsmtFinType2'] = modeling_data['BsmtFinType2'].fillna('None')
modeling_data['BsmtCond'] = modeling_data['BsmtCond'].fillna('None')
modeling_data['BsmtQual'] = modeling_data['BsmtQual'].fillna('None')
```

**17~18. MasVnrArea** & **MasVnrType**
- Description: : Square feet & Type of Masonry veneer area
- Processing: 'NaN'은 Masonry veneer가 없는 집이라 판단하고 'None'로 대체

```python
modeling_data['MasVnrArea'] = modeling_data['MasVnrArea'].fillna('None')
modeling_data['MasVnrType'] = modeling_data['MasVnrType'].fillna('None')
```

**19. Electrical**
- Description: : Electrical system
- Processing: 'NaN'은 Electrical system이 없는 집이라 판단하여 'None'로 대체

```python
modeling_data['Electrical'] = modeling_data['Electrical'].fillna('None')
```

#### Check still is there any variables contained missing values

```python
modeling_data_null_df = modeling_data.isnull().sum()
modeling_data_null_df = pd.DataFrame(modeling_data_null_df, columns=['Missing Value Count'])
modeling_data_null_df['Missing Value Ratio'] = modeling_data_null_df['Missing Value Count']/train.shape[0]
modeling_data_null_df = modeling_data_null_df[modeling_data_null_df['Missing Value Count']!=0]
modeling_data_null_df.sort_values('Missing Value Count',ascending=False,inplace=True)
modeling_data_null_df
```
