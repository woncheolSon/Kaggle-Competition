```python
fig, ax = plt.subplots(6,3, figsize = (20,20))

for idx, n in enumerate(real_vars):
    if n == 'SalePrice':
        continue
    sns.regplot(x=n, y='SalePrice', data=train, ax = ax[idx//3,idx%3])
    ax[idx//3, idx%3].set(yticklabels=[])
    ax[idx//3, idx%3].set(xticklabels=[])
```
